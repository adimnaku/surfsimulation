```python
import gym_duckietown
from gym_duckietown.simulator import Simulator
import matplotlib.pyplot as plt
import numpy as np
import math

import duckietown_world

env = Simulator(
        seed=123, # random seed
        map_name="straigh_road_obj",
        max_steps=500001, # we don't want the gym to reset itself
        domain_rand=0,
        camera_width=640,
        camera_height=480,
        accept_start_angle_deg=1,#at close to straight
        full_transparency=True,
        distortion=True,
    )  

for obj in env.objects:
    obj.velocity = -0.1
    # obj.angle = 0

dt = 0
sp1 = 0.1

while dt < 100:
    
    lane_pose = env.get_lane_pos2(env.cur_pos, env.cur_angle)
    distance_to_road_center = lane_pose.dist
    angle_from_straight_in_rads = lane_pose.angle_rad

    k_p = 10
    k_d = 1
    
    steering = k_p*distance_to_road_center + k_d*angle_from_straight_in_rads # TODO: You should overwrite this value
    
    obs, reward, done, misc = env.step([sp1, steering])
    
    dt = dt + 1
   
    if(sp1 < 0):
        sp1 = 0
    
    env.render()

    if done:
        if reward < 0:
            print('*** CRASHED ***')
        break

x = []
right_wheel = []
dist = []
left_wheel = []
angle = []
velocity = []
penalty = []
x_pos = []
z_pos = []
accel_list = []
dt = 0

figure, axis = plt.subplots(2, 3)
plt.subplots_adjust(left=0.1, bottom=0.01, right=0.9, top=0.9, wspace=0.9, hspace=0.9)

#initial parameters for the car
env.angle = 0
speed = 0.2
INITIAL_DISTANCE = 30
TIME_STEP = 1
obj_dist = INITIAL_DISTANCE
ACCEL_CONSTANT = 0
accel = math.pow(speed, 2) /(-2.0 * obj_dist)

# The speed is a value between [0, 1] (which corresponds to a real speed between 0m/s and 1.2m/s)
# If this value is too high it will crash in no time, try to set it to 1.

other_vel = 0

for obj in env.objects:
    obj.velocity = 0.02
    obj.angle = 0

while dt < 500:
    
    lane_pose = env.get_lane_pos2(env.cur_pos, env.cur_angle)
    distance_to_road_center = lane_pose.dist
    angle_from_straight_in_rads = lane_pose.angle_rad
    k_p = 10
    k_d = 1
    steering = k_p*distance_to_road_center + k_d*angle_from_straight_in_rads 

    ###### Getting info to be printed on the graphs
    obs, reward, done, misc = env.step([speed, steering])
    
    info = misc["Simulator"]
    wheel_vels = info["wheel_velocities"]
    pos = info["cur_pos"]
    
    x.append(dt)
    angle.append(info["cur_angle"])
    velocity.append(info["robot_speed"])
    penalty.append(info["proximity_penalty"])
    x_pos.append(pos[0])
    z_pos.append(pos[2])
    left_wheel.append(wheel_vels[1])
    right_wheel.append(wheel_vels[0])
    
    dt = dt + 1
    
    obj_dist -= speed * TIME_STEP + 0.5 * accel * math.pow(TIME_STEP, 2)
    speed = speed + accel * TIME_STEP
    
    if(dt > 105 and obj_dist > 0):
        rel_vel = speed + other_vel
        accel = math.pow(rel_vel, 2) /(-2.0 * obj_dist)
    
    dist.append(obj_dist)
    accel_list.append(accel)

    if(speed < 0):
        speed = 0
    
    env.render()

    if done:
        if reward < 0:
            print('*** CRASHED ***')
        break

axis[0][0].scatter(x, angle)
axis[0][0].set_title("Angle vs time")
axis[0][0].set_xlabel("time")
axis[0][0].set_ylabel("angle")

axis[0][2].scatter(x, velocity)
axis[0][2].set_title("Velocity vs time")
axis[0][2].set_xlabel("time")
axis[0][2].set_ylabel("velocity")

axis[0][1].scatter(dist, velocity)
axis[0][1].set_title("Velocity vs distance")
axis[0][1].set_xlabel("distance")
axis[0][1].set_ylabel("velocity")

axis[1][0].scatter(x, right_wheel)
axis[1][0].set_title("Right Wheel Velocity vs time")
axis[1][0].set_xlabel("time")
axis[1][0].set_ylabel("right wheel velocity")

axis[1][1].scatter(x, accel_list)
axis[1][1].set_title("Accelerations vs time")
axis[1][1].set_xlabel("time")
axis[1][1].set_ylabel("acceleration")

axis[1][2].scatter(x, left_wheel)
axis[1][2].set_title("Left Wheel Velocity vs time")
axis[1][2].set_xlabel("time")
axis[1][2].set_ylabel("left wheel velocity")

#axis[1][2].scatter(x_pos, z_pos)$
#axis[1][2].set_title("X Pos vs Z Pos")
#axis[1][2].set_xlabel("x position")
#axis[1][2].set_ylabel("z position")

#plt.scatter(x, angle, c="blue")
#plt.title("How Angle Changes Over Time")
#plt.xlabel("time")
#plt.ylabel("curr_angle");

#Chasing Car satsy

plt.show()
```
